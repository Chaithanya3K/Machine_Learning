By default, indexing in pandas starts at 0. We can change the index according to our convenience using the set_index() method, which takes parameters such as the column name to be used as the new index (and an optional inplace=True to apply changes directly). To return to the default numeric index, we use the reset_index() method, which also has an inplace parameter (default False). Setting inplace=True will reset the index within the same DataFrame without needing reassignment.

```python
df.set_index("PassengerId",inplace=True)
df.head()
```
